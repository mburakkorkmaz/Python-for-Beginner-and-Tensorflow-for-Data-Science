```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizi1 = np.linspace(0,10,20)
```

```python
numpyDizi1
```

```python
numpyDizi2 = numpyDizi1 ** 2
```

```python
(benimFigur,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizi1,numpyDizi2,"g")

benimEksen.plot(numpyDizi2,numpyDizi1,"b")
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizi1,numpyDizi1 +2 , color="red",linewidth=8)
yeniEksen.plot(numpyDizi1,numpyDizi1 +3 , color = "yellow", linewidth=6.0)

yeniEksen.plot(numpyDizi1,numpyDizi1+4, color="green", linestyle="-.")
yeniEksen.plot(numpyDizi1,numpyDizi1+5, color="green", linestyle=":")
yeniEksen.plot(numpyDizi1,numpyDizi1+6, color="green", linestyle="--")
```
