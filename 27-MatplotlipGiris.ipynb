```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.tight_layout()
```

```python
yasListesi = [10,20,30,40,50,60,70,75,80,85]
kiloListesi = [20,60,80,85,87,69,67,56,59,73]
```

```python
yasListesi = np.array(yasListesi)
```

```python
kiloListesi = np.array(kiloListesi)
```

```python
plt.plot(yasListesi,kiloListesi, "b") # red r , green g ,blue b, yellow y
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilo ve Yaş Grafiği")
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi1
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g.-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"b*-")
```

```python
myFigure = plt.figure()
```

```python
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X-Ekseni")
figureAxes.set_ylabel("Y-Ekseni")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X-Ekseni")
eksen1.set_ylabel("Y-Ekseni")
eksen1.set_title("Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X-Ekseni s")
eksen2.set_ylabel("Y-Ekseni s")
eksen2.set_title("Küçük Grafik Başlığı")
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)
# benimEksenler.plot(numpyDizisi1,numpyDizisi2,"b")
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g.-")
    eksen.set_xlabel("X-Eksen")
plt.tight_layout()
```

```python
yeniFigure = plt.Figure()
yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi2,label = "numpyDizisi ** 2")
yeniEksen.legend(loc=6)
```
